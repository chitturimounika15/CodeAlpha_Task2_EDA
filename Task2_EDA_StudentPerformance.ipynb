## Unique Values

```python
df.nunique()
```

## Average Scores

```python
df[['math score','reading score','writing score']].mean()
```

## Maximum Scores

```python
df[['math score','reading score','writing score']].max()
```

## Minimum Scores

```python
df[['math score','reading score','writing score']].min()
```

## Gender-wise Performance Analysis

```python
df.groupby('gender')[['math score','reading score','writing score']].mean()
```

Female students performed better in Reading and Writing.

Male students performed better in Mathematics.

## Correlation Analysis

```python
df[['math score','reading score','writing score']].corr()
```

Reading and Writing scores have a very strong positive correlation.

## Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

sns.heatmap(
    df[['math score','reading score','writing score']].corr(),
    annot=True,
    cmap='Blues'
)

plt.title("Correlation Heatmap")
plt.show()
```
